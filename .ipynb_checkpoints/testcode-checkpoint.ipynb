```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
import cdflib
from cdflib import cdfepoch
from datetime import date

ROOT = Path("data")       # CDF
CHOICE = "POD"            # "POD" or "ACC"
YEAR = 2020
THE_DAY = date(2020,2,16) # 

def pick_cdf_with_day(root, choice, year, the_day):
    targets = sorted(p for p in root.rglob("*.cdf")
                     if choice.upper() in p.name.upper() and f"{year}" in p.name)
    for p in targets:
        c = cdflib.CDF(str(p))
        # time UTCtz-aware
        t = pd.to_datetime(cdfepoch.to_datetime(c.varget("time")), utc=True)
        # CDF the_day 
        if (pd.Index(t.date) == the_day).any():
            return p
    return None

def load_raw_day_df(path, the_day):
    c = cdflib.CDF(str(path))
    t = pd.to_datetime(cdfepoch.to_datetime(c.varget("time")), utc=True)
    df = pd.DataFrame(index=pd.DatetimeIndex(t, name="time"))
    for v in ("altitude","density","latitude","longitude","local_solar_time"):
        try:
            arr = np.asarray(c.varget(v)).astype(float)
        except Exception:
            continue
        # FILLVAL  NaN "D""E"
        atts = c.varattsget(v) or {}
        fv = atts.get("FILLVAL")
        if isinstance(fv, (list, tuple, np.ndarray)) and len(fv)==1: fv = fv[0]
        if isinstance(fv, (bytes, bytearray)): fv = fv.decode()
        if isinstance(fv, str): fv = fv.replace("D","E")
        try:
            fv = float(fv)
            arr = np.where(arr == fv, np.nan, arr)
        except Exception:
            pass
        arr = np.where(arr > 1e30, np.nan, arr)
        df[v] = arr
    # 
    dmask = (df.index.date == the_day)
    return df.loc[dmask].sort_index()

cdf_path = pick_cdf_with_day(ROOT, CHOICE, YEAR, THE_DAY)
assert cdf_path is not None, "CDF"

raw = load_raw_day_df(cdf_path, THE_DAY)
print("CDF file:", cdf_path)
print("rows:", len(raw), "cols:", list(raw.columns))

# CDF
fig, ax = plt.subplots(2,1,figsize=(12,6), sharex=True)
if "altitude" in raw:
    ax[0].plot(raw.index, raw["altitude"]/1000)
    ax[0].set_ylabel("Altitude [km]")
if "density" in raw:
    ax[1].plot(raw.index, raw["density"])
    ax[1].set_ylabel("Density")
ax[0].set_title(f"{CHOICE} raw-from-CDF on {THE_DAY}")
ax[1].set_xlabel("UTC")
plt.tight_layout(); plt.show()
```

```python
# ===  ===
import os, calendar, pickle, numpy as np, pandas as pd
from pathlib import Path
import cdflib
from cdflib import cdfepoch
from datetime import date

# === ===
OUTDIR = (Path.cwd() / "outputs").resolve()
OUTDIR.mkdir(parents=True, exist_ok=True)
print("[CWD]", Path.cwd())
print("[OUTDIR]", OUTDIR, " writable=", os.access(OUTDIR, os.W_OK))

# === CDF===
def cdf_day_table(path: Path, the_day: date,
                  cols=("time","altitude","density","latitude","longitude","local_solar_time")) -> pd.DataFrame:
    c = cdflib.CDF(str(path))
    # time
    t = pd.to_datetime(cdfepoch.to_datetime(c.varget("time")), utc=True)
    df = pd.DataFrame(index=pd.DatetimeIndex(t, name="time"))

    want = [c for c in cols if c != "time"]
    all_vars = set(c.cdf_info()["rVariables"] + c.cdf_info()["zVariables"])
    for v in want:
        if v not in all_vars:
            continue
        arr = np.asarray(c.varget(v)).astype(float)
        atts = c.varattsget(v) or {}
        fv = atts.get("FILLVAL")
        if isinstance(fv, (list, tuple, np.ndarray)) and len(fv)==1: fv = fv[0]
        if isinstance(fv, (bytes, bytearray)): fv = fv.decode()
        if isinstance(fv, str): fv = fv.replace("D","E")
        try:
            fv = float(fv); arr = np.where(arr == fv, np.nan, arr)
        except Exception:
            pass
        arr = np.where(arr > 1e30, np.nan, arr)
        if arr.ndim == 1 and len(arr) == len(df):
            df[v] = arr
    out = df.loc[df.index.date == the_day].copy().reset_index()
    final_cols = [c for c in cols if c in out.columns]
    return out[final_cols]

# === DNSpickle===
def dns_day_table(dns_pickle_path: str, the_day: date,
                  cols=("time","altitude","density","latitude","longitude","local_solar_time","T")) -> pd.DataFrame:
    DNS = pickle.load(open(dns_pickle_path, "rb"))
    YEAR = int(dns_pickle_path.split("_")[-1].split(".")[0])  # : DNS_POD_2020.pkl  2020
    doy = (pd.Timestamp(the_day) - pd.Timestamp(f"{YEAR}-01-01")).days + 1
    if doy not in DNS:
        raise KeyError(f"{the_day} (DOY={doy})  {dns_pickle_path} ")
    df = DNS[doy]["data"].copy().reset_index()  # DatetimeIndex -> 
    df.rename(columns={"index":"time"}, inplace=True)
    final_cols = [c for c in cols if c in df.columns]
    return df[final_cols]

# ===  ===
def safe_save_table(df: pd.DataFrame, basename: str):
    csv_path = (OUTDIR / f"{basename}.csv").resolve()
    xlsx_path = (OUTDIR / f"{basename}.xlsx").resolve()
    df.to_csv(csv_path, index=False)
    try:
        # xlsx  openpyxl 
        import openpyxl  # type: ignore
        df.to_excel(xlsx_path, index=False)
        wrote_xlsx = True
    except Exception as e:
        wrote_xlsx = False
        print(f"[WARN] Excel: {e}")

    # 
    assert csv_path.exists(), f"CSV: {csv_path}"
    print(f"[OK] CSV: {csv_path}  size={csv_path.stat().st_size} bytes")
    if wrote_xlsx and xlsx_path.exists():
        print(f"[OK] XLSX: {xlsx_path}  size={xlsx_path.stat().st_size} bytes")
    # 
    print(df.head(3))

# =====    =====
# 1) DNS
THE_DAY = date(2020, 2, 16)
try:
    tbl_pod = dns_day_table("DNS_POD_2020.pkl", THE_DAY)
    safe_save_table(tbl_pod, f"dns_pod_{THE_DAY.isoformat()}")
except Exception as e:
    print("[DNS->TABLE ERROR]", e)

try:
    tbl_acc = dns_day_table("DNS_ACC_2020.pkl", THE_DAY)
    safe_save_table(tbl_acc, f"dns_acc_{THE_DAY.isoformat()}")
except Exception as e:
    print("[DNS->TABLE ERROR]", e)

# 2) CDF / 
# cdf_path = Path("data/.../CDF.cdf")
# try:
#     tbl_cdf = cdf_day_table(cdf_path, THE_DAY)
#     safe_save_table(tbl_cdf, f"cdf_{THE_DAY.isoformat()}")
# except Exception as e:
#     print("[CDF->TABLE ERROR]", e)

# ===  outputs/  ===
print("\n[LIST outputs/]")
for p in sorted(OUTDIR.glob("*")):
    print(" -", p.name, p.stat().st_size, "bytes")
```

```python
# === CDF -> CSV  (CDFInfo) ===
# : "by_day" or "by_file"
MODE = "by_day"

from pathlib import Path
import numpy as np
import pandas as pd
import cdflib
from cdflib import cdfepoch

# ===  ===
ROOT = Path("data")  # CDF
OUTDIR = (Path.cwd() / "outputs_cdf").resolve()
OUTDIR.mkdir(parents=True, exist_ok=True)

#  None 
CHOICE_FILTER = None   # "POD"  "ACC"  None
YEAR_FILTER   = None   # : 2020  None

# -------------  -------------
def _safe_float(val):
    if val is None: return None
    if isinstance(val, (list, tuple, np.ndarray)) and len(val) == 1:
        val = val[0]
    if isinstance(val, (bytes, bytearray)):
        val = val.decode("utf-8", errors="ignore")
    if isinstance(val, str):
        val = val.replace("D", "E")
    try:
        return float(val)
    except Exception:
        return None

def _guess_choice(path: Path) -> str:
    s = path.name.upper()
    if "POD" in s: return "POD"
    if "ACC" in s: return "ACC"
    return "UNK"

def _get_vars(info):
    """cdf_info()  dict  CDFInfo """
    if isinstance(info, dict):
        rvars = info.get("rVariables", [])
        zvars = info.get("zVariables", [])
    else:
        rvars = getattr(info, "rVariables", [])
        zvars = getattr(info, "zVariables", [])
    return rvars, zvars

def _find_time_var(c: cdflib.cdfread.CDF) -> str:
    """time Swarm: 'time' """
    info = c.cdf_info()
    rvars, zvars = _get_vars(info)
    vars_all = list(rvars) + list(zvars)
    for cand in ("time", "TIME", "Epoch", "EPOCH", "Timestamp", "TT2000"):
        if cand in vars_all:
            return cand
    # : zVariables 
    for v in zvars:
        try:
            pd.to_datetime(cdfepoch.to_datetime(c.varget(v)), utc=True)
            return v
        except Exception:
            continue
    raise KeyError("time ")

def read_cdf_all_vars(path: Path, use_valid_range: bool = True) -> pd.DataFrame:
    """
    CDFtimeUTC, tz-aware
    time  1/2 DataFrame 
    2 var_0, var_1, ... 
    """
    c = cdflib.CDF(str(path))
    info = c.cdf_info()
    _, zvars = _get_vars(info)

    time_var = _find_time_var(c)
    t = pd.to_datetime(cdfepoch.to_datetime(c.varget(time_var)), utc=True)
    idx = pd.DatetimeIndex(t, name="time")

    cols = {}
    for v in zvars:
        if v == time_var:
            continue
        try:
            arr = np.asarray(c.varget(v))
        except Exception:
            continue  # 

        # time  len(time)
        if arr.ndim == 0 or arr.shape[0] != len(idx):
            continue

        # 
        try:
            arr = arr.astype(float, copy=False)
        except Exception:
            continue

        #  FILLVAL / VALIDMIN/MAX 
        atts = c.varattsget(v) or {}
        fill = _safe_float(atts.get("FILLVAL"))
        vmin = _safe_float(atts.get("VALIDMIN")) if use_valid_range else None
        vmax = _safe_float(atts.get("VALIDMAX")) if use_valid_range else None

        def _nanize(X):
            if fill is not None:
                X = np.where(X == fill, np.nan, X)
            X = np.where(X > 1e30, np.nan, X)  # 
            if vmin is not None:
                X = np.where(X < vmin, np.nan, X)
            if vmax is not None:
                X = np.where(X > vmax, np.nan, X)
            return X

        if arr.ndim == 1:
            cols[v] = _nanize(arr)
        elif arr.ndim == 2:
            k = arr.shape[1]
            X = _nanize(arr)
            for j in range(k):
                cols[f"{v}_{j}"] = X[:, j]
        # 3

    df = pd.DataFrame(cols, index=idx).sort_index()
    return df

def save_csv(df: pd.DataFrame, path: Path):
    path.parent.mkdir(parents=True, exist_ok=True)
    df.reset_index().to_csv(path, index=False)
    print(f"[OK] saved: {path}  rows={len(df)}")

# -------------  -------------
files = sorted(ROOT.rglob("*.cdf"))
if CHOICE_FILTER:
    files = [p for p in files if CHOICE_FILTER.upper() in p.name.upper()]
if YEAR_FILTER:
    files = [p for p in files if str(YEAR_FILTER) in p.name]

print(f"[INFO] Found {len(files)} CDF files under {ROOT} (MODE={MODE})")

if MODE == "by_file":
    for p in files:
        choice = _guess_choice(p).lower()
        df = read_cdf_all_vars(p)
        if YEAR_FILTER:
            df = df[df.index.year == YEAR_FILTER]
        out = OUTDIR / f"cdf_{choice}_{p.stem}.csv"
        save_csv(df, out)

elif MODE == "by_day":
    day_buckets = {"POD": {}, "ACC": {}, "UNK": {}}
    for p in files:
        choice = _guess_choice(p)  # "POD" / "ACC" / "UNK"
        df = read_cdf_all_vars(p)
        if YEAR_FILTER:
            df = df[df.index.year == YEAR_FILTER]
        for d, g in df.groupby(df.index.date):
            key = str(d)
            day_buckets[choice].setdefault(key, []).append(g)

    total_saved = 0
    for choice, bucket in day_buckets.items():
        for date_str, frames in sorted(bucket.items()):
            day_df = pd.concat(frames).sort_index()
            if not day_df.index.is_unique:
                day_df = day_df[~day_df.index.duplicated(keep="last")]
            out = OUTDIR / f"cdf_{choice.lower()}_{date_str}.csv"
            save_csv(day_df, out)
            total_saved += 1
    print(f"[DONE] Saved {total_saved} daily CSV files into {OUTDIR}")

else:
    raise ValueError("MODE must be 'by_day' or 'by_file'")
```

```python
import glob
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
FIGDIR = (Path.cwd() / "figs").resolve()
FIGDIR.mkdir(parents=True, exist_ok=True)

# 1) CSV
def load_day_csv(path):
    df = pd.read_csv(path, parse_dates=["time"]).set_index("time").sort_index()
    return df

def load_day_pair(day_str):
    """YYYY-MM-DD  POD/ACC CSVOK"""
    base = Path("outputs_cdf")
    pod = base / f"cdf_pod_{day_str}.csv"
    acc = base / f"cdf_acc_{day_str}.csv"
    P = load_day_csv(pod) if pod.exists() else None
    A = load_day_csv(acc) if acc.exists() else None
    return P, A

# 2) 
def plot_day_alt_den(day_str, resample="1min", log_density=False, save=True):
    P, A = load_day_pair(day_str)
    if P is None and A is None:
        print(f"[WARN] no CSV for {day_str}")
        return
    if resample:
        if P is not None: P = P.resample(resample).mean()
        if A is not None: A = A.resample(resample).mean()

    fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

    # km
    if P is not None and "altitude" in P:
        axes[0].plot(P.index, P["altitude"]/1000, label="POD", linestyle="--")
    if A is not None and "altitude" in A:
        axes[0].plot(A.index, A["altitude"]/1000, label="ACC", alpha=0.8)
    axes[0].set_ylabel("Altitude [km]"); axes[0].set_title(f"Altitude on {day_str}"); axes[0].legend()

    # 
    if P is not None and "density" in P:
        axes[1].plot(P.index, P["density"], label="POD density", linestyle="--")
    if A is not None and "density" in A:
        axes[1].plot(A.index, A["density"], label="ACC density", alpha=0.8)
    axes[1].set_ylabel("Density"); axes[1].set_title(f"Density on {day_str}"); axes[1].legend()
    if log_density:
        axes[1].set_yscale("log")

    axes[-1].set_xlabel("UTC")
    plt.tight_layout()
    if save:
        out = FIGDIR / f"day_{day_str}_alt_den.png"
        plt.savefig(out, dpi=180, bbox_inches="tight"); print("[OK] saved:", out)
    plt.show()

# 3) 
def plot_month_altitude(month_str, which="both", resample="5min", save=True):
    """
    month_str: '2020-02' 
    which: 'pod'/'acc'/'both'
    """
    base = Path("outputs_cdf")
    yy, mm = month_str.split("-")
    paths_p = sorted(glob.glob(str(base / f"cdf_pod_{yy}-{mm}-*.csv")))
    paths_a = sorted(glob.glob(str(base / f"cdf_acc_{yy}-{mm}-*.csv")))

    def concat(paths):
        if not paths: return None
        frames = [pd.read_csv(p, parse_dates=["time"]).set_index("time") for p in paths]
        df = pd.concat(frames).sort_index()
        return df

    P = concat(paths_p) if which in ("pod","both") else None
    A = concat(paths_a) if which in ("acc","both") else None
    if P is None and A is None:
        print(f"[WARN] no CSV for {month_str} ({which})"); return

    if resample:
        if P is not None: P = P.resample(resample).mean()
        if A is not None: A = A.resample(resample).mean()

    fig, ax = plt.subplots(figsize=(12, 4.5))
    if P is not None and "altitude" in P:
        ax.plot(P.index, P["altitude"]/1000, label="POD")
    if A is not None and "altitude" in A:
        ax.plot(A.index, A["altitude"]/1000, label="ACC", alpha=0.85)
    ax.set_ylabel("Altitude [km]"); ax.set_title(f"Monthly Altitude ({month_str})"); ax.legend()
    ax.set_xlabel("UTC"); plt.tight_layout()
    if save:
        out = FIGDIR / f"month_{month_str}_{which}_alt.png"
        plt.savefig(out, dpi=180, bbox_inches="tight"); print("[OK] saved:", out)
    plt.show()

# 4)  or 
def plot_month_heatmap(month_str, which="pod", var="density", bin_minutes=5, log10=False, save=True):
    """
    month_str: '2020-02'
    which: 'pod' or 'acc'
    var: 'density' or 'altitude'
    """
    base = Path("outputs_cdf")
    yy, mm = month_str.split("-")
    paths = sorted(glob.glob(str(base / f"cdf_{which}_{yy}-{mm}-*.csv")))
    if not paths:
        print(f"[WARN] no CSV for {month_str} ({which})"); return
    frames = [pd.read_csv(p, parse_dates=["time"]).set_index("time")[[var]] for p in paths if Path(p).exists()]
    df = pd.concat(frames).sort_index()
    # 5
    df = df.resample(f"{bin_minutes}min").mean()
    # =day, =
    w = df.copy()
    w["day"] = df.index.day
    w["tod"] = df.index.strftime("%H:%M")
    pivot = w.pivot_table(index="day", columns="tod", values=var, aggfunc="mean").sort_index()
    Z = pivot.values
    if log10 and var == "density":
        Z = np.where(Z>0, np.log10(Z), np.nan)

    fig, ax = plt.subplots(figsize=(12, 5))
    im = ax.imshow(Z, aspect="auto", interpolation="nearest", origin="upper")
    ax.set_title(f"{which.upper()} {var} heatmap ({month_str})  bin={bin_minutes}min" + (" [log10]" if log10 and var=="density" else ""))
    ax.set_ylabel("Day of month"); ax.set_xlabel("Time (UTC)")
    cols = pivot.columns.to_list()
    ticks = [0, len(cols)//4, len(cols)//2, 3*len(cols)//4, len(cols)-1]
    labels = [cols[i] for i in ticks]
    ax.set_xticks(ticks); ax.set_xticklabels(labels, rotation=0)
    ax.set_yticks(np.arange(pivot.shape[0])); ax.set_yticklabels(pivot.index)
    cbar = plt.colorbar(im, ax=ax); cbar.set_label(f"{var}"+(" (log10)" if log10 and var=="density" else ""))
    plt.tight_layout()
    if save:
        out = FIGDIR / f"heatmap_{which}_{var}_{month_str}_bin{bin_minutes}.png"
        plt.savefig(out, dpi=180, bbox_inches="tight"); print("[OK] saved:", out)
    plt.show()

# 5) POD/ACC 1+ 
def make_compare_day(day_str, freq="1min", save_table=True, save_plot=True):
    P, A = load_day_pair(day_str)
    if P is None or A is None:
        print(f"[WARN] need both POD and ACC for {day_str}"); return
    P1 = P.resample(freq).mean().add_suffix("_pod")
    A1 = A.resample(freq).mean().add_suffix("_acc")
    cmp = P1.join(A1, how="outer")
    if save_table:
        out = FIGDIR / f"compare_{day_str}_{freq}.csv"
        cmp.to_csv(out); print("[OK] saved table:", out)
    fig, axes = plt.subplots(1,2,figsize=(12,4))
    if "altitude_pod" in cmp and "altitude_acc" in cmp:
        axes[0].plot(cmp.index, (cmp["altitude_acc"]-cmp["altitude_pod"])/1000)
        axes[0].set_title("Altitude ACC - POD [km]"); axes[0].set_ylabel("km")
    if "density_pod" in cmp and "density_acc" in cmp:
        ok = cmp[["density_pod","density_acc"]].dropna()
        axes[1].scatter(ok["density_pod"], ok["density_acc"], s=8, alpha=0.5)
        axes[1].set_xlabel("POD density"); axes[1].set_ylabel("ACC density"); axes[1].set_title("Density correlation")
    plt.tight_layout()
    if save_plot:
        outp = FIGDIR / f"compare_{day_str}_{freq}.png"
        plt.savefig(outp, dpi=180, bbox_inches="tight"); print("[OK] saved plot:", outp)
    plt.show()
```

```python
import math, glob
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
FIGDIR = (Path.cwd() / "figs").resolve()
FIGDIR.mkdir(parents=True, exist_ok=True)

# ---- CSV ----
def _load_day_csv(path: Path) -> pd.DataFrame | None:
    if not path.exists():
        return None
    df = pd.read_csv(path, parse_dates=["time"]).set_index("time").sort_index()
    return df

def load_day_pair(day_str: str, source: str = "cdf"):
    """
    source: 'cdf' -> outputs_cdf/cdf_*.csv
            'dns' -> outputs/dns_*.csv  DNSCSV
    """
    base = Path("outputs_cdf") if source == "cdf" else Path("outputs")
    prefix = "cdf" if source == "cdf" else "dns"
    pod = _load_day_csv(base / f"{prefix}_pod_{day_str}.csv")
    acc = _load_day_csv(base / f"{prefix}_acc_{day_str}.csv")
    return pod, acc

# ----  ----
def plot_all_vars_by_time(
    day_str: str,
    *,
    source: str = "cdf",          # 'cdf' or 'dns'
    which: str = "both",          # 'pod' / 'acc' / 'both'
    resample: str | None = None,  # : "1min" / None()
    page_size: int = 8,           # 1
    log_vars: tuple[str,...] = ("density",),  # OK
    kmify_altitude: bool = True,  # altitudekm
    save: bool = True
):
    P, A = load_day_pair(day_str, source=source)
    if which.lower() == "pod":
        A = None
    elif which.lower() == "acc":
        P = None
    if P is None and A is None:
        print(f"[WARN] {day_str}: {source} POD/ACCCSV")
        return

    # 
    if resample:
        if P is not None: P = P.resample(resample).mean()
        if A is not None: A = A.resample(resample).mean()

    # timeNaN
    def numeric_cols(df: pd.DataFrame) -> list[str]:
        if df is None: return []
        return [c for c, dt in df.dtypes.items() if np.issubdtype(dt, np.number)]

    cols = sorted(set(numeric_cols(P)) | set(numeric_cols(A)))

    if not cols:
        print(f"[WARN] {day_str}: ")
        return

    # page_size
    def is_log_var(name: str) -> bool:
        # log_vars
        return any(name.startswith(pref) for pref in log_vars)

    pages = math.ceil(len(cols) / page_size)
    fnames = []

    for pidx in range(pages):
        chunk = cols[pidx*page_size : (pidx+1)*page_size]
        fig, axes = plt.subplots(len(chunk), 1, figsize=(12, 3.2*len(chunk)), sharex=True)
        if len(chunk) == 1:
            axes = [axes]

        plotted_any = False

        for ax, col in zip(axes, chunk):
            # 
            sP = P[col] if (P is not None and col in P.columns) else None
            sA = A[col] if (A is not None and col in A.columns) else None

            # NaN
            if (sP is None or sP.dropna().empty) and (sA is None or sA.dropna().empty):
                ax.set_visible(False)
                continue

            # altitudekm
            label_unit = ""
            if kmify_altitude and col.startswith("altitude"):
                if sP is not None: sP = sP / 1000.0
                if sA is not None: sA = sA / 1000.0
                label_unit = " [km]"

            # 
            if sP is not None:
                ax.plot(sP.index, sP.values, linestyle="--", label="POD")
                plotted_any = True
            if sA is not None:
                ax.plot(sA.index, sA.values, alpha=0.85, label="ACC")
                plotted_any = True

            # 
            if is_log_var(col):
                ax.set_yscale("log")
                label_unit += " (log)"

            ax.set_ylabel(f"{col}{label_unit}")
            ax.legend(loc="best")
            ax.grid(True, alpha=0.2)

        axes[-1].set_xlabel("UTC")
        fig.suptitle(f"{day_str}  variables vs time  (source={source}, which={which}, resample={resample or 'native'})",
                     y=0.995, fontsize=12)
        plt.tight_layout(rect=[0, 0, 1, 0.98])

        if plotted_any and save:
            out = FIGDIR / f"allvars_{day_str}_{source}_{which}_page{pidx+1}.png"
            plt.savefig(out, dpi=180, bbox_inches="tight")
            print("[OK] saved:", out)
            fnames.append(out)

        plt.show()

    if not fnames:
        print("[INFO] NaN")
```

```python
# CDFCSVPOD/ACC1
plot_all_vars_by_time("2020-02-16", source="cdf", which="both", resample="1min")

# DNS30s/10s
plot_all_vars_by_time("2020-02-16", source="dns", which="both", resample=None)

# ACC110
plot_all_vars_by_time("2020-03-05", source="cdf", which="acc", resample="1min",
                      page_size=10, log_vars=("density",))
```

```python
def _superscript_pow(p: int) -> str:
    # e.g., -13 -> ""
    trans = str.maketrans("-0123456789", "")
    return str(p).translate(trans)

def plot_day_alt_den(day_str, resample="1min", log_density=False,
                     save=True, density_unit_pow: int | None = None):
    P, A = load_day_pair(day_str)
    if P is None and A is None:
        print(f"[WARN] no CSV for {day_str}")
        return
    if resample:
        if P is not None: P = P.resample(resample).mean()
        if A is not None: A = A.resample(resample).mean()

    # density 10^(-p)  = 10^(+|p|) 
    d_scale = (10.0 ** (-density_unit_pow)) if density_unit_pow is not None else 1.0
    d_unit_label = f" [10{_superscript_pow(density_unit_pow)}]" if density_unit_pow is not None else ""

    fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

    # km
    if P is not None and "altitude" in P:
        axes[0].plot(P.index, P["altitude"]/1000, label="POD", linestyle="--")
    if A is not None and "altitude" in A:
        axes[0].plot(A.index, A["altitude"]/1000, label="ACC", alpha=0.8)
    axes[0].set_ylabel("Altitude [km]"); axes[0].set_title(f"Altitude on {day_str}"); axes[0].legend()

    # 10^-13 
    if P is not None and "density" in P:
        axes[1].plot(P.index, P["density"]*d_scale, label="POD density", linestyle="--")
    if A is not None and "density" in A:
        axes[1].plot(A.index, A["density"]*d_scale, label="ACC density", alpha=0.8)
    axes[1].set_ylabel(f"Density{d_unit_label}")
    axes[1].set_title(f"Density on {day_str}")
    axes[1].legend()
    if log_density:
        axes[1].set_yscale("log")

    axes[-1].set_xlabel("UTC")
    plt.tight_layout()
    if save:
        out = FIGDIR / f"day_{day_str}_alt_den.png"
        plt.savefig(out, dpi=180, bbox_inches="tight"); print("[OK] saved:", out)
    plt.show()
```

```python
def plot_month_heatmap(month_str, which="pod", var="density",
                       bin_minutes=5, log10=False, save=True,
                       density_unit_pow: int | None = None):
    base = Path("outputs_cdf")
    yy, mm = month_str.split("-")
    paths = sorted(glob.glob(str(base / f"cdf_{which}_{yy}-{mm}-*.csv")))
    if not paths:
        print(f"[WARN] no CSV for {month_str} ({which})"); return
    frames = [pd.read_csv(p, parse_dates=["time"]).set_index("time")[[var]]
              for p in paths if Path(p).exists()]
    df = pd.concat(frames).sort_index()
    df = df.resample(f"{bin_minutes}min").mean()

    # density 
    d_unit_label = ""
    if var.startswith("density") and density_unit_pow is not None:
        df[var] = df[var] * (10.0 ** (-density_unit_pow))
        d_unit_label = f" [10{_superscript_pow(density_unit_pow)}]"

    w = df.copy()
    w["day"] = df.index.day
    w["tod"] = df.index.strftime("%H:%M")
    pivot = w.pivot_table(index="day", columns="tod", values=var, aggfunc="mean").sort_index()
    Z = pivot.values
    if log10 and var.startswith("density"):
        Z = np.where(Z > 0, np.log10(Z), np.nan)

    fig, ax = plt.subplots(figsize=(12, 5))
    im = ax.imshow(Z, aspect="auto", interpolation="nearest", origin="upper")
    title = f"{which.upper()} {var} heatmap ({month_str})  bin={bin_minutes}min"
    if log10 and var.startswith("density"):
        cbar_label = f"{var}{d_unit_label} (log10)"
    else:
        cbar_label = f"{var}{d_unit_label}"
    ax.set_title(title)
    ax.set_ylabel("Day of month"); ax.set_xlabel("Time (UTC)")
    cols = pivot.columns.to_list()
    ticks = [0, len(cols)//4, len(cols)//2, 3*len(cols)//4, len(cols)-1]
    labels = [cols[i] for i in ticks]
    ax.set_xticks(ticks); ax.set_xticklabels(labels, rotation=0)
    ax.set_yticks(np.arange(pivot.shape[0])); ax.set_yticklabels(pivot.index)
    cbar = plt.colorbar(im, ax=ax); cbar.set_label(cbar_label)
    plt.tight_layout()
    if save:
        out = FIGDIR / f"heatmap_{which}_{var}_{month_str}_bin{bin_minutes}.png"
        plt.savefig(out, dpi=180, bbox_inches="tight"); print("[OK] saved:", out)
    plt.show()
```

```python
def plot_all_vars_by_time(day_str: str, *,
    source: str = "cdf", which: str = "both",
    resample: str | None = None, page_size: int = 8,
    log_vars: tuple[str,...] = ("density",),
    kmify_altitude: bool = True, save: bool = True,
    density_unit_pow: int | None = None):
    P, A = load_day_pair(day_str, source=source)
    if which.lower() == "pod": A = None
    elif which.lower() == "acc": P = None
    if P is None and A is None:
        print(f"[WARN] {day_str}: {source} POD/ACCCSV"); return
    if resample:
        if P is not None: P = P.resample(resample).mean()
        if A is not None: A = A.resample(resample).mean()

    d_scale = (10.0 ** (-density_unit_pow)) if density_unit_pow is not None else 1.0
    d_unit_label = f" [10{_superscript_pow(density_unit_pow)}]" if density_unit_pow is not None else ""

    def numeric_cols(df):
        if df is None: return []
        return [c for c, dt in df.dtypes.items() if np.issubdtype(dt, np.number)]
    cols = sorted(set(numeric_cols(P)) | set(numeric_cols(A)))
    if not cols:
        print(f"[WARN] {day_str}: "); return

    import math
    pages = math.ceil(len(cols) / page_size)
    for pidx in range(pages):
        chunk = cols[pidx*page_size : (pidx+1)*page_size]
        fig, axes = plt.subplots(len(chunk), 1, figsize=(12, 3.2*len(chunk)), sharex=True)
        if len(chunk) == 1: axes = [axes]
        plotted_any = False
        for ax, col in zip(axes, chunk):
            sP = P[col] if (P is not None and col in P.columns) else None
            sA = A[col] if (A is not None and col in A.columns) else None
            if (sP is None or sP.dropna().empty) and (sA is None or sA.dropna().empty):
                ax.set_visible(False); continue

            label_unit = ""
            # altitudekm
            if kmify_altitude and col.startswith("altitude"):
                if sP is not None: sP = sP / 1000.0
                if sA is not None: sA = sA / 1000.0
                label_unit = " [km]"
            # density10^pow
            if col.startswith("density") and density_unit_pow is not None:
                if sP is not None: sP = sP * d_scale
                if sA is not None: sA = sA * d_scale
                label_unit = d_unit_label

            if sP is not None:
                ax.plot(sP.index, sP.values, linestyle="--", label="POD"); plotted_any = True
            if sA is not None:
                ax.plot(sA.index, sA.values, alpha=0.85, label="ACC"); plotted_any = True

            if any(col.startswith(pref) for pref in log_vars):
                ax.set_yscale("log")
                if label_unit: label_unit += " (log)"

            ax.set_ylabel(f"{col}{label_unit}")
            ax.legend(loc="best"); ax.grid(True, alpha=0.2)
        axes[-1].set_xlabel("UTC")
        fig.suptitle(f"{day_str}  variables vs time  (source={source}, which={which}, resample={resample or 'native'})",
                     y=0.995, fontsize=12)
        plt.tight_layout(rect=[0,0,1,0.98])
        if plotted_any and save:
            out = FIGDIR / f"allvars_{day_str}_{source}_{which}_page{pidx+1}.png"
            plt.savefig(out, dpi=180, bbox_inches="tight"); print("[OK] saved:", out)
        plt.show()
```
